## Procurar numero de suites

```python
def extrair_numero_suite(descricao):
    # Use expressão regular para encontrar o número de suítes
    match = re.search(r'(\d+)\s*su[ií]te[s]*', descricao, re.IGNORECASE)
    
    if match:
        return int(match.group(1))
    else:
        return 0

df_imob['Número de Suítes'] = df_imob['Descricao'].apply(extrair_numero_suite)
```

```python
df_imob['Número de Suítes'].value_counts()
```

```python
df_imob = df_imob.loc[df_imob['Número de Suítes'] <= 10,:]
```

## Encontrar Sobrado Duplex triplex

```python
def verificar_sobrado(row):
    descricao = row['Descricao']
    titulo = row['Titulo']
    
    # Use expressão regular para verificar a presença de "Sobrado" ou derivações
    regex_pattern = r'sobrad[oaos]*|duplex|triplex|casa de 2 andares'
    
    if re.search(regex_pattern, descricao, re.IGNORECASE) or re.search(regex_pattern, titulo, re.IGNORECASE):
        return True
    else:
        return False

df_imob['Sobrado'] = df_imob.apply(verificar_sobrado, axis=1)
```

```python
import requests

def geocode_address(address, api_key):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}"
    response = requests.get(url)
    data = response.json()
    
    if data['status'] == "OK":
        location = data['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        return None, None

# Exemplo:
address = "Avenida José da Silva Sé - de 2000 ao fim - lado par Condomínio Residencial Parque da Liberdade I - São José do Rio Preto, SP"
api_key = "AIzaSyBN6nHVeWl8xQk6Zxokq7XOjb2EcEHF3PU"
latitude, longitude = geocode_address(address, api_key)
print(latitude, longitude)
```

```python
import time

def geocode_unique_addresses(df, api_key, column_name='endereco'):
    # Extract unique addresses
    unique_addresses = df[column_name].dropna().unique()

    # Geocode the unique addresses
    mapping = {}
    for address in unique_addresses:
        lat, lng = geocode_address(address, api_key)
        mapping[address] = (lat, lng)
        
        # Pause for a short time to avoid hitting the rate limit
        time.sleep(0.1)

    # Apply the mapping to the dataframe
    df['latitude'] = df[column_name].map(lambda x: mapping.get(x, (None, None))[0])
    df['longitude'] = df[column_name].map(lambda x: mapping.get(x, (None, None))[1])
    
    return df

# Exemplo:
df_region = geocode_unique_addresses(df1, api_key)
```

```python
from shapely.geometry import Point
from shapely.geometry import Polygon

# Dicionário para armazenar os polígonos das regiões
regioes_poligonos = {regiao: Polygon(coords) for regiao, coords in regioes_coords_atualizadas.items()}

def determinar_regiao(row, regioes_poligonos):
    ponto = Point(row['latitude'], row['longitude'])
    
    for regiao, poligono in regioes_poligonos.items():
        if poligono.contains(ponto):
            return regiao
    return None

# Adicionar coluna de região ao dataframe
df_region['regiao'] = df_region.apply(lambda row: determinar_regiao(row, regioes_poligonos), axis=1)
```

## Neural Networks

```python
X_train_neural = X_train.copy()
X_val_neural = X_val.copy()
```

```python
X_train.columns
```

```python
bool_columns = X_train.select_dtypes(include='bool')
bool_columns = bool_columns.columns
```

```python
X_train_neural[bool_columns] = X_train[bool_columns].astype(int)
X_val_neural[bool_columns] = X_val_neural[bool_columns].astype(int)
```

```python
X_train_neural = X_train_neural.astype(float)
X_val_neural = X_val_neural.astype(float)
```

```python
#!pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
# Crie o modelo da rede neural
model = keras.Sequential([
    keras.layers.BatchNormalization(),
    keras.layers.Dense(64, activation='relu', kernel_initializer='he_normal',
                       input_shape=(X_train.shape[1],)),
    keras.layers.Dropout(0.1),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(32, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dropout(0.1),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(16, activation='relu', kernel_initializer='he_normal'),
    keras.layers.Dense(1)  # Camada de saída
])

optimizer = keras.optimizers.Adam(learning_rate=1e-2)
# Compile o modelo
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=[keras.metrics.MeanSquaredError()])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=25, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=5, min_lr=1e-6)


history = model.fit(
    X_train_neural,
    y_train, 
    validation_data=(X_val_neural, y_val),
    epochs=500,
    callbacks = [early_stopping, reduce_lr]
)
```

```python
y_pred = model.predict(X_val_neural)
y_pred = y_pred.reshape(-1,)
```

```python
print(f'MAE: {mean_absolute_error(y_pred, y_val)}')
print(f'MAPE: {mean_absolute_percentage_error(y_pred, y_val)}')
print(f'MSE: {mean_squared_error(y_pred, y_val)}')
print(f'RMSE: {mean_squared_error(y_pred, y_val, squared=False)}')
```

## Fine Tunning

```python
import optuna
import xgboost as xgb
```

```python
def objective(trial):
    params = {
        "objective": "reg:squarederror",
        "n_estimators": 1000,
        "verbosity": 0,
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.03, log=True),
        "max_depth": trial.suggest_int("max_depth", 6, 20),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.7, 1.0),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
    }

    model = xgb.XGBRegressor(**params)
    model.fit(X_train, y_train, verbose=False)
    predictions = model.predict(X_val)
    rmse = mean_squared_error(y_val, predictions, squared=False)
    return rmse
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)
```

```python
print('Best hyperparameters:', study.best_params)
print('Best RMSE:', study.best_value)
```

```python
import requests

def geocode_address(address, api_key):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}"
    response = requests.get(url)
    data = response.json()
    
    if data['status'] == "OK":
        location = data['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        return None, None
```

```python
from shapely.geometry import Point
from shapely.geometry import Polygon

# Dicionário para armazenar os polígonos das regiões
regioes_poligonos = {regiao: Polygon(coords) for regiao, coords in regioes_coords_atualizadas.items()}

def determinar_regiao(row, regioes_poligonos):
    ponto = Point(row['latitude'], row['longitude'])
    
    for regiao, poligono in regioes_poligonos.items():
        if poligono.contains(ponto):
            return regiao
    return None

# Adicionar coluna de região ao dataframe
df_dados['regiao'] = df_dados.apply(lambda row: determinar_regiao(row, regioes_poligonos), axis=1)
```

```python
regioes_coords_atualizadas = {
    'Talhados':
[(-20.680000, -49.266000), 
(-20.770000, -49.279000),
#bordas com represa 
(-20.769000, -49.287000),
(-20.765000, -49.298000),
(-20.765000, -49.301000),
(-20.763000, -49.306000),
(-20.761000, -49.306000),
(-20.765000, -49.316000),
(-20.760000, -49.317000),
(-20.762000, -49.321000),
(-20.760000, -49.323000),
(-20.766000, -49.328000),
(-20.764000, -49.329000),
(-20.774000, -49.349000),
(-20.773000, -49.351000),
 #fim borda represa
(-20.751000, -49.341000),
(-20.750000, -49.340000),
(-20.748000, -49.340000),
(-20.742000, -49.338000),
(-20.665000, -49.341000),
(-20.680000, -49.266000)

],
'Represa':[
(-20.770000, -49.279000),
(-20.843000, -49.254000),
#inicio limites schimitdt
(-20.843000, -49.264000),
(-20.843000, -49.281000),
(-20.828000, -49.289000),
(-20.839000, -49.305000),
(-20.838000, -49.309000),
#fim limites schimitd
#inicio vila toninho  
(-20.838000, -49.309000),
(-20.826000, -49.307000),
(-20.819000, -49.324000),
(-20.821000, -49.329000),
(-20.828000, -49.335000),
(-20.827000, -49.339000),
(-20.826000, -49.343000),
(-20.828000, -49.343000),
(-20.825000, -49.348000),
(-20.817000, -49.350000),
(-20.811000, -49.357000),
#fim vila toninho
#começo borda centro
(-20.811000, -49.357000),
(-20.781000, -49.354000),
#fim borda centro
#unica borda com bosque    
(-20.776000, -49.353000),
#fim borda bosque    
#bordas com talhados    
(-20.773000, -49.351000),
(-20.774000, -49.349000),
(-20.764000, -49.329000),
(-20.766000, -49.328000),
(-20.760000, -49.323000),
(-20.762000, -49.321000),
(-20.760000, -49.317000),
(-20.765000, -49.316000),
(-20.761000, -49.306000),
(-20.763000, -49.306000),
(-20.765000, -49.301000),
(-20.765000, -49.298000),
(-20.770000, -49.279000),
#(-20.769000, -49.286000),
#(-20.770000, -49.282000),
],
    'Vila Toninho':[
#limites com represa        
(-20.811000, -49.357000),
(-20.817000, -49.350000),
(-20.825000, -49.348000),
(-20.828000, -49.343000),
(-20.826000, -49.343000),
(-20.827000, -49.339000),
(-20.828000, -49.335000),
(-20.821000, -49.329000),
(-20.819000, -49.324000),
(-20.826000, -49.307000),
(-20.838000, -49.309000),
#fim limites represa    
#Inicio schimidt        
(-20.838000, -49.309000),        
(-20.846000, -49.308000),
(-20.848000, -49.313000),
(-20.846000, -49.314000),
(-20.846000, -49.315000),
(-20.844000, -49.315000),
(-20.846000, -49.323000),
(-20.845000, -49.326000),
(-20.852000, -49.330000),
(-20.853000, -49.328000),
(-20.860000, -49.335000),
(-20.863000, -49.332000),
(-20.876000, -49.349000),
#fim limite schmidt
#começo HB
(-20.876000, -49.349000),
(-20.876000, -49.354000),
(-20.874000, -49.355000),
(-20.870000, -49.353000),
(-20.855000, -49.363000),
(-20.853000, -49.365000),
(-20.845000, -49.358000),
(-20.841000, -49.352000),
(-20.831000, -49.362000),
#fim limite hb
#inicio centro
(-20.831000, -49.362000),
(-20.826000, -49.359000),
(-20.824000, -49.358000),
(-20.811000, -49.357000),
#fim centro        
],

    'Schimidt': [
#inicio represa
(-20.843000, -49.264000),
(-20.843000, -49.281000),
(-20.828000, -49.289000),
(-20.839000, -49.305000),
(-20.838000, -49.309000),
#fim represa
#inicio toninho
(-20.838000, -49.309000),        
(-20.846000, -49.308000),
(-20.848000, -49.313000),
(-20.846000, -49.314000),
(-20.846000, -49.315000),
(-20.844000, -49.315000),
(-20.846000, -49.323000),
(-20.845000, -49.326000),
(-20.852000, -49.330000),
(-20.853000, -49.328000),
(-20.860000, -49.335000),
(-20.863000, -49.332000),
(-20.876000, -49.349000),
#fim toninho
#inicio HB   
(-20.876000, -49.349000),
(-20.881000, -49.345000),
(-20.882000, -49.347000),
(-20.884000, -49.348000),
(-20.885000, -49.352000),
(-20.890000, -49.355000),
(-20.905000, -49.341000),
(-20.906000, -49.347000),
(-20.910000, -49.347000),
(-20.912000, -49.349000),
(-20.920000, -49.351000),
(-20.931000, -49.348000),
#fim hb
(-20.936000, -49.344000),
(-20.883000, -49.274000),
(-20.864000, -49.255000),
(-20.854000, -49.255000),
(-20.843000, -49.264000),
    ],
    'HB':[
#inicio Schimidt
(-20.876000, -49.349000),
(-20.881000, -49.345000),
(-20.882000, -49.347000),
(-20.884000, -49.348000),
(-20.885000, -49.352000),
(-20.890000, -49.355000),
(-20.905000, -49.341000),
(-20.906000, -49.347000),
(-20.910000, -49.347000),
(-20.912000, -49.349000),
(-20.920000, -49.351000),
(-20.931000, -49.348000),
#fim schimidt
(-20.919000, -49.415000),
(-20.883000, -49.444000),
(-20.816000, -49.449000),
#inicio cidade crianças
(-20.816000, -49.449000),
(-20.817000, -49.438000),
(-20.822000, -49.424000),
(-20.823000, -49.419000),
#fim cidade crianças
#inicio centro
(-20.823000, -49.419000),
(-20.825000, -49.406000),
(-20.824000, -49.396000),
(-20.824000, -49.374000),
(-20.826000, -49.367000),
(-20.831000, -49.362000),
#fim centro
#inicio vila toninho
(-20.831000, -49.362000),
(-20.841000, -49.352000),
(-20.845000, -49.358000),
(-20.853000, -49.365000),
(-20.855000, -49.363000),
(-20.870000, -49.353000),
(-20.874000, -49.355000),
(-20.876000, -49.354000),
(-20.876000, -49.349000),
        
    ],

  'Cidade das Crianças':[
#inicio HB
(-20.816000, -49.449000),
(-20.817000, -49.438000),
(-20.822000, -49.424000),
(-20.823000, -49.419000),
#fim hb
#inicio centro
(-20.823000, -49.419000), 
(-20.816000, -49.411000),
(-20.810000, -49.412000),
(-20.810000, -49.409000),
(-20.807000, -49.410000),
(-20.807000, -49.409000),
(-20.806000, -49.409000),
(-20.801000, -49.400000),
(-20.791000, -49.388000),
#fim centro
#inicio pinheirinho
(-20.791000, -49.388000),
(-20.788000, -49.389000),
(-20.784000, -49.393000),
(-20.782000, -49.398000),
(-20.779000, -49.402000),
(-20.785000, -49.408000),
(-20.785000, -49.407000),
(-20.788000, -49.409000),
(-20.787000, -49.410000),
(-20.789000, -49.412000),
(-20.788000, -49.414000),
(-20.783000, -49.416000),
(-20.782000, -49.416000),
(-20.780000, -49.423000),
(-20.780000, -49.425000),
(-20.779000, -49.425000),
(-20.779000, -49.427000),
(-20.776000, -49.427000),
(-20.777000, -49.429000),
(-20.779000, -49.430000),
(-20.779000, -49.428000),
(-20.783000, -49.428000),
(-20.784000, -49.432000),
(-20.782000, -49.435000),
(-20.783000, -49.446000),
(-20.786000, -49.445000),
(-20.791000, -49.458000),
#fim pinheirinho
(-20.816000, -49.449000),
      
  ],

    'Pinheirinho':[
#inicio cidade das crianças
(-20.791000, -49.388000),
(-20.788000, -49.389000),
(-20.784000, -49.393000),
(-20.782000, -49.398000),
(-20.779000, -49.402000),
(-20.785000, -49.408000),
(-20.785000, -49.407000),
(-20.788000, -49.409000),
(-20.787000, -49.410000),
(-20.789000, -49.412000),
(-20.788000, -49.414000),
(-20.783000, -49.416000),
(-20.782000, -49.416000),
(-20.780000, -49.423000),
(-20.780000, -49.425000),
(-20.779000, -49.425000),
(-20.779000, -49.427000),
(-20.776000, -49.427000),
(-20.777000, -49.429000),
(-20.779000, -49.430000),
(-20.779000, -49.428000),
(-20.783000, -49.428000),
(-20.784000, -49.432000),
(-20.782000, -49.435000),
(-20.783000, -49.446000),
(-20.786000, -49.445000),
(-20.791000, -49.458000),
#fim cidade das crianças
(-20.788000, -49.460000),
(-20.685000, -49.469000),
(-20.685000, -49.462000),
(-20.702000, -49.423000),
(-20.722000, -49.425000),
#inicio CEU
(-20.722000, -49.425000),
(-20.727000, -49.426000),
(-20.728000, -49.428000),
(-20.741000, -49.423000),
(-20.741000, -49.421000),
(-20.760000, -49.416000),
(-20.760000, -49.414000),
(-20.763000, -49.413000),
(-20.766000, -49.412000),
(-20.766000, -49.410000),
(-20.767000, -49.408000),
(-20.765000, -49.403000),
(-20.766000, -49.400000),
(-20.766000, -49.398000),
(-20.766000, -49.393000),
(-20.768000, -49.393000),
(-20.769000, -49.388000),
(-20.767000, -49.389000),
(-20.767000, -49.385000),
(-20.770000, -49.385000),
(-20.778000, -49.386000),
(-20.779000, -49.382000),
#fim ceu
#inicio BOSQUE
(-20.779000, -49.382000),
(-20.780000, -49.382000),
(-20.782000, -49.380000),
#fim bosque
#inicio centro
(-20.782000, -49.380000),
(-20.784000, -49.381000),
(-20.785000, -49.379000),
(-20.789000, -49.381000),
(-20.789000, -49.383000),
(-20.791000, -49.388000),

    ],

    'CEU':[
#inicio pinheiro
(-20.722000, -49.425000),
(-20.727000, -49.426000),
(-20.728000, -49.428000),
(-20.741000, -49.423000),
(-20.741000, -49.421000),
(-20.760000, -49.416000),
(-20.760000, -49.414000),
(-20.763000, -49.413000),
(-20.766000, -49.412000),
(-20.766000, -49.410000),
(-20.767000, -49.408000),
(-20.765000, -49.403000),
(-20.766000, -49.400000),
(-20.766000, -49.398000),
(-20.766000, -49.393000),
(-20.768000, -49.393000),
(-20.769000, -49.388000),
(-20.767000, -49.389000),
(-20.767000, -49.385000),
(-20.770000, -49.385000),
(-20.778000, -49.386000),
(-20.779000, -49.382000),
#fim pinheiro
#inicio bosque
(-20.779000, -49.382000),
(-20.773000, -49.378000),
(-20.766000, -49.377000),
(-20.759000, -49.374000),
(-20.755000, -49.375000),
(-20.749000, -49.375000),
(-20.749000, -49.381000),
(-20.743000, -49.380000),
(-20.725000, -49.381000),
(-20.718000, -49.386000),
(-20.709000, -49.387000),
#fim bosque
(-20.706000, -49.393000),
(-20.704000, -49.393000),
(-20.703000, -49.401000),
(-20.722000, -49.425000),
    ],
    'BOSQUE':[
#inicio ceu
(-20.779000, -49.382000),
(-20.773000, -49.378000),
(-20.766000, -49.377000),
(-20.759000, -49.374000),
(-20.755000, -49.375000),
(-20.749000, -49.375000),
(-20.749000, -49.381000),
(-20.743000, -49.380000),
(-20.725000, -49.381000),
(-20.718000, -49.386000),
(-20.709000, -49.387000),
#fim ceu
(-20.724000, -49.369000),
(-20.717000, -49.362000),
(-20.679000, -49.349000),
(-20.671000, -49.365000),
(-20.635000, -49.354000),
(-20.645000, -49.345000),
(-20.665000, -49.341000),
#inicio talhados
(-20.665000, -49.341000),
(-20.742000, -49.338000),
(-20.748000, -49.340000),
(-20.750000, -49.340000),
(-20.751000, -49.341000),
(-20.773000, -49.351000),
#fim talhados
#inicio represa
(-20.773000, -49.351000),
#fim represa
#inicio centro
(-20.781000, -49.354000),
(-20.791000, -49.364000),
(-20.789000, -49.366000),
(-20.787000, -49.367000),
(-20.787000, -49.368000),
(-20.780000, -49.368000),
(-20.779000, -49.374000),
(-20.781000, -49.378000),
(-20.782000, -49.380000),
#fim centro
#inicio pinheirinho
(-20.782000, -49.380000),
#fim pinheironho
(-20.779000, -49.382000)
    ],

    'Centro':[
#cidade das crianças
(-20.823000, -49.419000), 
(-20.816000, -49.411000),
(-20.810000, -49.412000),
(-20.810000, -49.409000),
(-20.807000, -49.410000),
(-20.807000, -49.409000),
(-20.806000, -49.409000),
(-20.801000, -49.400000),
(-20.791000, -49.388000),
#pinheirinho
(-20.791000, -49.388000),
(-20.789000, -49.383000),
(-20.789000, -49.381000),
(-20.785000, -49.379000),
(-20.784000, -49.381000),
(-20.782000, -49.380000),
#bosque
(-20.782000, -49.380000),
(-20.781000, -49.378000),
(-20.779000, -49.374000),
(-20.780000, -49.368000),
(-20.787000, -49.368000),
(-20.787000, -49.367000),
(-20.789000, -49.366000),
(-20.791000, -49.364000),
(-20.781000, -49.354000),
#represa
(-20.781000, -49.354000),
(-20.811000, -49.357000),
#vila toninho
(-20.811000, -49.357000),
(-20.824000, -49.358000),
(-20.826000, -49.359000),
(-20.831000, -49.362000),
#HB
(-20.831000, -49.362000),
(-20.826000, -49.367000),
(-20.824000, -49.374000),
(-20.824000, -49.396000),
(-20.825000, -49.406000),
(-20.823000, -49.419000),

    ]
    

}
ordem_desejada = [
    'quartos', 'banheiros', 'vagas_de_garagem', 'area',
    'academia', 'aquecimento', 'ar_condicionado', 'area_de_servico',
    'armarios_na_cozinha', 'armarios_no_quarto', 'banheiro_no_quarto',
    'churrasqueira', 'internet', 'mobiliado', 'piscina',
    'porteiro_24h', 'quarto_de_servico', 'salao_de_festas',
    'tv_a_cabo', 'varanda', 'area_murada', 'condominio_fechado',
    'elevador', 'permitido_animais', 'portao_eletronico', 'portaria',
    'seguranca_24h', 'numero_de_suites', 'sobrado',
    'latitude', 'longitude', 'regiao_BOSQUE', 'regiao_CEU',
    'regiao_Centro', 'regiao_Cidade das Crianças', 'regiao_HB',
    'regiao_Pinheirinho', 'regiao_Represa', 'regiao_Schimidt', 'regiao_Talhados',
    'regiao_Vila Toninho', 'tipo_apartamento', 'tipo_casa'
]

regioes_possiveis = ['regiao_BOSQUE',
  'regiao_CEU',
  'regiao_Centro',
  'regiao_BOSQUE',
  'regiao_CEU',
  'regiao_Cidade das Crianças',
  'regiao_HB',
  'regiao_Pinheirinho',
  'regiao_Represa',
  'regiao_Schimidt',
  'regiao_Talhados' ,
  'regiao_Vila Toninho' ]


import pandas as pd
from shapely.geometry import Point, Polygon
import requests

def geocode_address(address, api_key):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}"
    response = requests.get(url)
    data = response.json()
    
    if data['status'] == "OK":
        location = data['results'][0]['geometry']['location']
        return location['lat'], location['lng']
    else:
        return None, None

def determinar_regiao(row, regioes_poligonos):
    ponto = Point(row['latitude'], row['longitude'])
    
    for regiao, poligono in regioes_poligonos.items():
        if poligono.contains(ponto):
            return regiao
    return None

def prever_preco_imovel(input_features, api_key):
    
    
    # Crie um DataFrame a partir do input_features
    df_dados = pd.DataFrame(input_features, index=[0])

    # Geocode para obter latitude e longitude a partir do endereço
    latitude, longitude = geocode_address(df_dados['endereco'], api_key)
    df_dados['latitude'] = latitude
    df_dados['longitude'] = longitude

    # Dicionário para armazenar os polígonos das regiões
    regioes_poligonos = {regiao: Polygon(coords) for regiao, coords in regioes_coords_atualizadas.items()}

    # Adicionar coluna de região ao dataframe
    df_dados['regiao'] = df_dados.apply(lambda row: determinar_regiao(row, regioes_poligonos), axis=1)


    coluna_regiao = None  # Defina inicialmente como None
    
    # Determine a região com base na latitude e longitude e defina como True
    regiao = determinar_regiao(df_dados.iloc[0], regioes_poligonos)
    if regiao is not None:
        coluna_regiao = f"regiao_{regiao}"
        df_dados[coluna_regiao] = True

    # Preencha as outras colunas de região com False
    for coluna in regioes_possiveis:
        if coluna != coluna_regiao:
            df_dados[coluna] = False

    df_dados = df_dados.drop(['regiao','endereco'], axis=1)

    scaler = joblib.load('/home/edilson07/projects/rp_housing/notebooks/standscaler_v2.joblib')
    mms = joblib.load('/home/edilson07/projects/rp_housing/notebooks/minmax_v2.joblib')
    

    cols_standard = ['area']
    cols_mms = ['latitude', 'longitude']

    # Aplique as transformações nos dados
    df_dados[cols_standard] = scaler.transform(df_dados[cols_standard])
    df_dados[cols_mms] = mms.transform(df_dados[cols_mms])

    #ordenar
    df_dados = df_dados[ordem_desejada]

    # Faça a previsão usando o modelo
    full_xgb = joblib.load('/home/edilson07/projects/rp_housing/notebooks/xgboost_best.joblib')
    preco_imovel = full_xgb.predict(df_dados)

    return preco_imovel

# Exemplo de uso:
#input_features = {
    # Inclua suas características de entrada aqui
#}

```
